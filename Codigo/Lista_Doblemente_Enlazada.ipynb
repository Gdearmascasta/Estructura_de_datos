```python
# ==========================================
# Curso: Estructura de Datos
# Presentado por:
# Angel Vega Rodriguez - T00068186
# Germán De Armas - T00068765
# ==========================================

# Listas Doblemente Enlazadas

# Nodo con referencia al siguiente y anterior
class Nodo:
    def __init__(self, dato):
        self.dato = dato
        self.next = None
        self.prev = None


class ListaDoble:
    def __init__(self):
        self.inicio = None
        self.fin = None

    # Insertar al inicio
    def insertarInicio(self, x):
        n = Nodo(x)

        if self.inicio == None:
            self.inicio = n
            self.fin = n
        else:
            n.next = self.inicio
            self.inicio.prev = n
            self.inicio = n

    # Insertar al final
    def insertarFinal(self, x):
        n = Nodo(x)

        if self.fin == None:
            self.inicio = n
            self.fin = n
        else:
            self.fin.next = n
            n.prev = self.fin
            self.fin = n

    # Buscar elemento
    def buscar(self, x):
        aux = self.inicio

        while aux != None:
            if aux.dato == x:
                return True
            aux = aux.next

        return False

    # Eliminar elemento
    def eliminar(self, x):
        aux = self.inicio

        while aux != None and aux.dato != x:
            aux = aux.next

        if aux == None:
            return

        if aux.prev == None:
            self.inicio = aux.next
            if self.inicio != None:
                self.inicio.prev = None
        else:
            aux.prev.next = aux.next

        if aux.next != None:
            aux.next.prev = aux.prev
        else:
            self.fin = aux.prev

    # Mostrar lista
    def mostrar(self):
        aux = self.inicio
        while aux != None:
            print(aux.dato, end=" <-> ")
            aux = aux.next
        print("None")

    # Ordenar ascendente
    def ordenarAsc(self):
        cambio = True
        while cambio:
            cambio = False
            aux = self.inicio

            while aux != None and aux.next != None:
                if aux.dato > aux.next.dato:
                    temp = aux.dato
                    aux.dato = aux.next.dato
                    aux.next.dato = temp
                    cambio = True
                aux = aux.next

    # Ordenar descendente
    def ordenarDesc(self):
        cambio = True
        while cambio:
            cambio = False
            aux = self.inicio

            while aux != None and aux.next != None:
                if aux.dato < aux.next.dato:
                    temp = aux.dato
                    aux.dato = aux.next.dato
                    aux.next.dato = temp
                    cambio = True
                aux = aux.next
```

```python
# ===== PROGRAMA PRINCIPAL =====

lista = ListaDoble()

# Carga inicial
n = int(input("Cantidad de datos: "))

for i in range(n):
    lista.insertarFinal(int(input(f"Dato {i+1}: ")))

# Menú interactivo
while True:
    print("\n--- MENU LISTA DOBLE ---")
    print("1. Insertar al inicio")
    print("2. Insertar al final")
    print("3. Buscar")
    print("4. Eliminar")
    print("5. Mostrar")
    print("6. Ordenar ascendente")
    print("7. Ordenar descendente")
    print("0. Salir")

    op = int(input("Seleccione: "))

    if op == 1:
        lista.insertarInicio(int(input("Dato: ")))

    elif op == 2:
        lista.insertarFinal(int(input("Dato: ")))

    elif op == 3:
        print("Encontrado:", lista.buscar(int(input("Buscar: "))))

    elif op == 4:
        lista.eliminar(int(input("Eliminar: ")))

    elif op == 5:
        lista.mostrar()

    elif op == 6:
        lista.ordenarAsc()
        print("Lista ordenada ascendentemente")

    elif op == 7:
        lista.ordenarDesc()
        print("Lista ordenada descendentemente")

    elif op == 0:
        print("Fin del programa")
        break

    else:
        print("Opción inválida")
```
