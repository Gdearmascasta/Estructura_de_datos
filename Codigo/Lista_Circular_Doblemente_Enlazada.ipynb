```python
# ==========================================
# Curso: Estructura de Datos
# Presentado por:
# Angel Vega Rodriguez - T00068186
# Germán De Armas - T00068765
# ==========================================

# Lista Circular Doblemente Enlazada

class Nodo:
    def __init__(self, dato):
        self.dato = dato
        self.next = None
        self.prev = None


class ListaCircularDoble:
    def __init__(self):
        self.inicio = None

    # Insertar al inicio
    def insertarInicio(self, x):
        n = Nodo(x)

        if self.inicio == None:
            n.next = n
            n.prev = n
            self.inicio = n
        else:
            ultimo = self.inicio.prev

            n.next = self.inicio
            n.prev = ultimo

            ultimo.next = n
            self.inicio.prev = n

            self.inicio = n

    # Insertar al final
    def insertarFinal(self, x):
        if self.inicio == None:
            self.insertarInicio(x)
        else:
            n = Nodo(x)
            ultimo = self.inicio.prev

            ultimo.next = n
            n.prev = ultimo

            n.next = self.inicio
            self.inicio.prev = n

    # Insertar ordenado
    def insertarOrdenado(self, x):
        if self.inicio == None or x < self.inicio.dato:
            self.insertarInicio(x)
            return

        aux = self.inicio

        while aux.next != self.inicio and aux.next.dato < x:
            aux = aux.next

        n = Nodo(x)

        n.next = aux.next
        n.prev = aux

        aux.next.prev = n
        aux.next = n

    # Buscar
    def buscar(self, x):
        if self.inicio == None:
            return False

        aux = self.inicio

        while True:
            if aux.dato == x:
                return True
            aux = aux.next
            if aux == self.inicio:
                break

        return False

    # Eliminar
    def eliminar(self, x):
        if self.inicio == None:
            return

        aux = self.inicio

        while True:
            if aux.dato == x:
                if aux.next == aux:  # único nodo
                    self.inicio = None
                else:
                    aux.prev.next = aux.next
                    aux.next.prev = aux.prev

                    if aux == self.inicio:
                        self.inicio = aux.next
                return

            aux = aux.next
            if aux == self.inicio:
                break

    # Mostrar
    def mostrar(self):
        if self.inicio == None:
            print("Lista vacía")
            return

        aux = self.inicio

        while True:
            print(aux.dato, end=" <-> ")
            aux = aux.next
            if aux == self.inicio:
                break

        print("(inicio)")

    # Orden ascendente
    def ordenarAsc(self):
        if self.inicio == None:
            return

        cambio = True
        while cambio:
            cambio = False
            aux = self.inicio

            while True:
                sig = aux.next
                if sig != self.inicio and aux.dato > sig.dato:
                    aux.dato, sig.dato = sig.dato, aux.dato
                    cambio = True
                aux = aux.next
                if aux.next == self.inicio:
                    break

    # Orden descendente
    def ordenarDesc(self):
        if self.inicio == None:
            return

        cambio = True
        while cambio:
            cambio = False
            aux = self.inicio

            while True:
                sig = aux.next
                if sig != self.inicio and aux.dato < sig.dato:
                    aux.dato, sig.dato = sig.dato, aux.dato
                    cambio = True
                aux = aux.next
                if aux.next == self.inicio:
                    break
```

```python
lista = ListaCircularDoble()

n = int(input("Cantidad de datos: "))
for i in range(n):
    lista.insertarFinal(int(input("Dato: ")))

while True:
    print("\n--- MENU LISTA CIRCULAR DOBLE ---")
    print("1. Insertar inicio")
    print("2. Insertar final")
    print("3. Insertar ordenado")
    print("4. Buscar")
    print("5. Eliminar")
    print("6. Mostrar")
    print("7. Ordenar ascendente")
    print("8. Ordenar descendente")
    print("0. Salir")

    op = int(input())

    if op == 1:
        lista.insertarInicio(int(input()))
    elif op == 2:
        lista.insertarFinal(int(input()))
    elif op == 3:
        lista.insertarOrdenado(int(input()))
    elif op == 4:
        print(lista.buscar(int(input())))
    elif op == 5:
        lista.eliminar(int(input()))
    elif op == 6:
        lista.mostrar()
    elif op == 7:
        lista.ordenarAsc()
    elif op == 8:
        lista.ordenarDesc()
    elif op == 0:
        break
```
