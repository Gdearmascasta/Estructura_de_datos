```python
# ==========================================
# Curso: Estructura de Datos
# Presentado por:
# Angel Vega Rodriguez - T00068186
# Germán De Armas - T00068765
# ==========================================

# Pila implementada con arreglo

class Pila:
    def __init__(self, n):
        self.arr = [0] * n   # arreglo fijo
        self.top = -1        # indica la posición del tope
        self.n = n           # tamaño máximo

    # Insertar elemento (push)
    def push(self, x):
        if self.top < self.n - 1:
            self.top += 1
            self.arr[self.top] = x
        else:
            print("Overflow (pila llena)")

    # Eliminar elemento (pop)
    def pop(self):
        if self.top >= 0:
            val = self.arr[self.top]
            self.top -= 1
            return val
        else:
            print("Underflow (pila vacía)")

    # Ver elemento en el tope (peek)
    def peek(self):
        if self.top >= 0:
            return self.arr[self.top]
        else:
            print("Pila vacía")

    # Tamaño de la pila
    def size(self):
        return self.top + 1

    # Mostrar pila (de arriba hacia abajo)
    def mostrar(self):
        if self.top == -1:
            print("Pila vacía")
            return

        i = self.top
        while i >= 0:
            print(self.arr[i])
            i -= 1

    # Ordenar ascendente
    def ordenarAsc(self):
        for i in range(self.top + 1):
            j = 0
            while j < self.top:
                if self.arr[j] > self.arr[j + 1]:
                    temp = self.arr[j]
                    self.arr[j] = self.arr[j + 1]
                    self.arr[j + 1] = temp
                j += 1

    # Ordenar descendente
    def ordenarDesc(self):
        for i in range(self.top + 1):
            j = 0
            while j < self.top:
                if self.arr[j] < self.arr[j + 1]:
                    temp = self.arr[j]
                    self.arr[j] = self.arr[j + 1]
                    self.arr[j + 1] = temp
                j += 1
```

```python
# ===== PROGRAMA PRINCIPAL =====

n = int(input("Tamaño de la pila: "))
p = Pila(n)

# Carga inicial
for i in range(n):
    p.push(int(input(f"Dato {i+1}: ")))

# Menú
while True:
    print("\n--- MENU PILA (ARREGLO) ---")
    print("1. Push")
    print("2. Pop")
    print("3. Peek")
    print("4. Size")
    print("5. Mostrar")
    print("6. Ordenar ascendente")
    print("7. Ordenar descendente")
    print("0. Salir")

    op = int(input("Seleccione: "))

    if op == 1:
        p.push(int(input("Dato: ")))

    elif op == 2:
        print("Elemento eliminado:", p.pop())

    elif op == 3:
        print("Tope:", p.peek())

    elif op == 4:
        print("Cantidad:", p.size())

    elif op == 5:
        p.mostrar()

    elif op == 6:
        p.ordenarAsc()
        print("Pila ordenada ascendentemente")

    elif op == 7:
        p.ordenarDesc()
        print("Pila ordenada descendentemente")

    elif op == 0:
        print("Fin del programa")
        break

    else:
        print("Opción inválida")
```
