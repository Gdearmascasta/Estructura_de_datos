```python
# ==========================================
# Curso: Estructura de Datos
# Presentado por:
# Angel Vega Rodriguez - T00068186
# Germán De Armas - T00068765
# ==========================================

# Pila implementada usando lista enlazada

# Nodo de la pila
class Nodo:
    def __init__(self, dato):
        self.dato = dato
        self.next = None


class Pila:
    def __init__(self):
        self.top = None   # referencia al tope
        self.c = 0        # contador de elementos

    # Insertar elemento (push)
    def push(self, x):
        n = Nodo(x)
        n.next = self.top
        self.top = n
        self.c += 1

    # Eliminar elemento (pop)
    def pop(self):
        if self.top != None:
            val = self.top.dato
            self.top = self.top.next
            self.c -= 1
            return val
        else:
            print("Underflow (pila vacía)")

    # Ver el tope (peek)
    def peek(self):
        if self.top != None:
            return self.top.dato
        else:
            print("Pila vacía")

    # Tamaño de la pila
    def size(self):
        return self.c

    # Mostrar pila
    def mostrar(self):
        if self.top == None:
            print("Pila vacía")
            return

        aux = self.top
        while aux != None:
            print(aux.dato)
            aux = aux.next

    # Ordenar ascendente
    def ordenarAsc(self):
        if self.top == None:
            return

        cambio = True
        while cambio:
            cambio = False
            aux = self.top

            while aux.next != None:
                if aux.dato > aux.next.dato:
                    temp = aux.dato
                    aux.dato = aux.next.dato
                    aux.next.dato = temp
                    cambio = True
                aux = aux.next

    # Ordenar descendente
    def ordenarDesc(self):
        if self.top == None:
            return

        cambio = True
        while cambio:
            cambio = False
            aux = self.top

            while aux.next != None:
                if aux.dato < aux.next.dato:
                    temp = aux.dato
                    aux.dato = aux.next.dato
                    aux.next.dato = temp
                    cambio = True
                aux = aux.next
```

```python
# ===== PROGRAMA PRINCIPAL =====

p = Pila()

n = int(input("Cantidad de datos iniciales: "))

for i in range(n):
    p.push(int(input(f"Dato {i+1}: ")))

while True:
    print("\n--- MENU PILA (LISTA) ---")
    print("1. Push")
    print("2. Pop")
    print("3. Peek")
    print("4. Size")
    print("5. Mostrar")
    print("6. Ordenar ascendente")
    print("7. Ordenar descendente")
    print("0. Salir")

    op = int(input("Seleccione: "))

    if op == 1:
        p.push(int(input("Dato: ")))

    elif op == 2:
        print("Elemento eliminado:", p.pop())

    elif op == 3:
        print("Tope:", p.peek())

    elif op == 4:
        print("Cantidad:", p.size())

    elif op == 5:
        p.mostrar()

    elif op == 6:
        p.ordenarAsc()
        print("Pila ordenada ascendentemente")

    elif op == 7:
        p.ordenarDesc()
        print("Pila ordenada descendentemente")

    elif op == 0:
        print("Fin del programa")
        break

    else:
        print("Opción inválida")
```
