```python
# ==========================================
# Curso: Estructura de Datos
# Presentado por:
# Angel Vega Rodriguez - T00068186
# Germán De Armas - T00068765
# ==========================================

# Cola implementada con arreglo

class Cola:
    def __init__(self, n):
        self.arr = [0] * n   # arreglo fijo
        self.f = 0           # frente
        self.r = -1          # final
        self.n = n           # tamaño máximo

    # Insertar elemento (enqueue)
    def enqueue(self, x):
        if self.r < self.n - 1:
            self.r += 1
            self.arr[self.r] = x
        else:
            print("Overflow (cola llena)")

    # Eliminar elemento (dequeue)
    def dequeue(self):
        if self.f <= self.r:
            val = self.arr[self.f]
            self.f += 1
            return val
        else:
            print("Underflow (cola vacía)")

    # Ver frente (peek)
    def peek(self):
        if self.f <= self.r:
            return self.arr[self.f]
        else:
            print("Cola vacía")

    # Tamaño de la cola
    def size(self):
        return self.r - self.f + 1

    # Mostrar cola
    def mostrar(self):
        if self.f > self.r:
            print("Cola vacía")
            return

        i = self.f
        while i <= self.r:
            print(self.arr[i], end=" ")
            i += 1
        print()

    # Ordenar ascendente
    def ordenarAsc(self):
        for i in range(self.r + 1):
            j = 0
            while j < self.r:
                if self.arr[j] > self.arr[j + 1]:
                    temp = self.arr[j]
                    self.arr[j] = self.arr[j + 1]
                    self.arr[j + 1] = temp
                j += 1

    # Ordenar descendente
    def ordenarDesc(self):
        for i in range(self.r + 1):
            j = 0
            while j < self.r:
                if self.arr[j] < self.arr[j + 1]:
                    temp = self.arr[j]
                    self.arr[j] = self.arr[j + 1]
                    self.arr[j + 1] = temp
                j += 1
```

```python
# ===== PROGRAMA PRINCIPAL =====

n = int(input("Tamaño de la cola: "))
c = Cola(n)

# Carga inicial
for i in range(n):
    c.enqueue(int(input(f"Dato {i+1}: ")))

# Menú
while True:
    print("\n--- MENU COLA (ARREGLO) ---")
    print("1. Enqueue")
    print("2. Dequeue")
    print("3. Peek")
    print("4. Size")
    print("5. Mostrar")
    print("6. Ordenar ascendente")
    print("7. Ordenar descendente")
    print("0. Salir")

    op = int(input("Seleccione: "))

    if op == 1:
        c.enqueue(int(input("Dato: ")))

    elif op == 2:
        print("Elemento eliminado:", c.dequeue())

    elif op == 3:
        print("Frente:", c.peek())

    elif op == 4:
        print("Cantidad:", c.size())

    elif op == 5:
        c.mostrar()

    elif op == 6:
        c.ordenarAsc()
        print("Cola ordenada ascendentemente")

    elif op == 7:
        c.ordenarDesc()
        print("Cola ordenada descendentemente")

    elif op == 0:
        print("Fin del programa")
        break

    else:
        print("Opción inválida")
```
