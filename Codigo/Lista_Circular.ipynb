```python
# ==========================================
# Curso: Estructura de Datos
# Presentado por:
# Angel Vega Rodriguez - T00068186
# Germán De Armas - T00068765
# ==========================================

# Lista Circular

class Nodo:
    def __init__(self, dato):
        self.dato = dato
        self.next = None


class ListaCircular:
    def __init__(self):
        self.inicio = None

    # Insertar al inicio
    def insertarInicio(self, x):
        n = Nodo(x)

        if self.inicio == None:
            self.inicio = n
            n.next = n
        else:
            aux = self.inicio
            while aux.next != self.inicio:
                aux = aux.next

            n.next = self.inicio
            aux.next = n
            self.inicio = n

    # Insertar al final
    def insertarFinal(self, x):
        n = Nodo(x)

        if self.inicio == None:
            self.inicio = n
            n.next = n
        else:
            aux = self.inicio
            while aux.next != self.inicio:
                aux = aux.next

            aux.next = n
            n.next = self.inicio

    # Insertar ordenado
    def insertarOrdenado(self, x):
        if self.inicio == None:
            self.insertarInicio(x)
            return

        if x < self.inicio.dato:
            self.insertarInicio(x)
            return

        n = Nodo(x)
        aux = self.inicio

        while aux.next != self.inicio and aux.next.dato < x:
            aux = aux.next

        n.next = aux.next
        aux.next = n

    # Buscar elemento
    def buscar(self, x):
        if self.inicio == None:
            return False

        aux = self.inicio

        while True:
            if aux.dato == x:
                return True
            aux = aux.next
            if aux == self.inicio:
                break

        return False

    # Eliminar elemento
    def eliminar(self, x):
        if self.inicio == None:
            return

        aux = self.inicio
        prev = None

        # Caso: eliminar el inicio
        if self.inicio.dato == x:
            if self.inicio.next == self.inicio:
                self.inicio = None
                return

            while aux.next != self.inicio:
                aux = aux.next

            aux.next = self.inicio.next
            self.inicio = self.inicio.next
            return

        prev = self.inicio
        aux = self.inicio.next

        while aux != self.inicio and aux.dato != x:
            prev = aux
            aux = aux.next

        if aux.dato == x:
            prev.next = aux.next

    # Mostrar lista
    def mostrar(self):
        if self.inicio == None:
            print("Lista vacía")
            return

        aux = self.inicio

        while True:
            print(aux.dato, end=" -> ")
            aux = aux.next
            if aux == self.inicio:
                break

        print("(inicio)")

    # Ordenar ascendente
    def ordenarAsc(self):
        if self.inicio == None:
            return

        cambio = True

        while cambio:
            cambio = False
            aux = self.inicio

            while True:
                sig = aux.next
                if sig != self.inicio and aux.dato > sig.dato:
                    temp = aux.dato
                    aux.dato = sig.dato
                    sig.dato = temp
                    cambio = True
                aux = aux.next
                if aux.next == self.inicio:
                    break

    # Ordenar descendente
    def ordenarDesc(self):
        if self.inicio == None:
            return

        cambio = True

        while cambio:
            cambio = False
            aux = self.inicio

            while True:
                sig = aux.next
                if sig != self.inicio and aux.dato < sig.dato:
                    temp = aux.dato
                    aux.dato = sig.dato
                    sig.dato = temp
                    cambio = True
                aux = aux.next
                if aux.next == self.inicio:
                    break
```

```python
# ===== PROGRAMA PRINCIPAL =====

lista = ListaCircular()

n = int(input("Cantidad de datos: "))

for i in range(n):
    lista.insertarFinal(int(input(f"Dato {i+1}: ")))

while True:
    print("\n--- MENU LISTA CIRCULAR ---")
    print("1. Insertar inicio")
    print("2. Insertar final")
    print("3. Insertar ordenado")
    print("4. Buscar")
    print("5. Eliminar")
    print("6. Mostrar")
    print("7. Ordenar ascendente")
    print("8. Ordenar descendente")
    print("0. Salir")

    op = int(input("Seleccione: "))

    if op == 1:
        lista.insertarInicio(int(input("Dato: ")))

    elif op == 2:
        lista.insertarFinal(int(input("Dato: ")))

    elif op == 3:
        lista.insertarOrdenado(int(input("Dato: ")))

    elif op == 4:
        print("Encontrado:", lista.buscar(int(input("Buscar: "))))

    elif op == 5:
        lista.eliminar(int(input("Eliminar: ")))

    elif op == 6:
        lista.mostrar()

    elif op == 7:
        lista.ordenarAsc()
        print("Orden ascendente aplicado")

    elif op == 8:
        lista.ordenarDesc()
        print("Orden descendente aplicado")

    elif op == 0:
        print("Fin del programa")
        break

    else:
        print("Opción inválida")
```
