```python
# ==========================================
# Curso: Estructura de Datos
# Presentado por:
# Angel Vega Rodriguez - T00068186
# Germán De Armas - T00068765
# ==========================================

# Listas Enlazadas Simples


# Nodo de la lista (almacena dato y referencia al siguiente)
class Nodo:
    def __init__(self, dato):
        self.dato = dato
        self.next = None


# Lista enlazada simple
class ListaSimple:
    def __init__(self):
        self.inicio = None  # referencia al primer nodo

    # Inserta un nodo al inicio de la lista
    def insertarInicio(self, x):
        n = Nodo(x)
        n.next = self.inicio
        self.inicio = n

    # Inserta un nodo al final recorriendo la lista
    def insertarFinal(self, x):
        n = Nodo(x)

        if self.inicio == None:
            self.inicio = n
        else:
            aux = self.inicio
            while aux.next != None:
                aux = aux.next
            aux.next = n

    # Inserta manteniendo orden ascendente (requiere lista ordenada)
    def insertarOrdenado(self, x):
        n = Nodo(x)

        if self.inicio == None or x < self.inicio.dato:
            n.next = self.inicio
            self.inicio = n
        else:
            aux = self.inicio
            while aux.next != None and aux.next.dato < x:
                aux = aux.next

            n.next = aux.next
            aux.next = n

    # Busca un elemento en la lista
    def buscar(self, x):
        aux = self.inicio

        while aux != None:
            if aux.dato == x:
                return True
            aux = aux.next

        return False

    # Elimina un nodo por valor
    def eliminar(self, x):
        if self.inicio == None:
            return

        # Caso: eliminar primer nodo
        if self.inicio.dato == x:
            self.inicio = self.inicio.next
            return

        aux = self.inicio

        # Buscar nodo anterior al que se desea eliminar
        while aux.next != None and aux.next.dato != x:
            aux = aux.next

        # Ajustar enlaces
        if aux.next != None:
            aux.next = aux.next.next

    # Muestra la lista completa
    def mostrar(self):
        aux = self.inicio

        while aux != None:
            print(aux.dato, end=" -> ")
            aux = aux.next

        print("None")

    # Ordena la lista de forma ascendente
    def ordenarAsc(self):
        if self.inicio == None:
            return

        cambio = True
        while cambio:
            cambio = False
            aux = self.inicio

            while aux.next != None:
                if aux.dato > aux.next.dato:
                    temp = aux.dato
                    aux.dato = aux.next.dato
                    aux.next.dato = temp
                    cambio = True
                aux = aux.next

    # Ordena la lista de forma descendente
    def ordenarDesc(self):
        if self.inicio == None:
            return

        cambio = True
        while cambio:
            cambio = False
            aux = self.inicio

            while aux.next != None:
                if aux.dato < aux.next.dato:
                    temp = aux.dato
                    aux.dato = aux.next.dato
                    aux.next.dato = temp
                    cambio = True
                aux = aux.next
```

```python
# ===== PROGRAMA PRINCIPAL =====

lista = ListaSimple()

# Carga inicial de datos
n = int(input("Cantidad inicial de datos: "))

for i in range(n):
    lista.insertarFinal(int(input(f"Dato {i+1}: ")))

# Menú interactivo
while True:
    print("\n--- MENU LISTA SIMPLE ---")
    print("1. Insertar inicio")
    print("2. Insertar final")
    print("3. Insertar ordenado")
    print("4. Buscar")
    print("5. Eliminar")
    print("6. Mostrar")
    print("7. Ordenar ascendente")
    print("8. Ordenar descendente")
    print("0. Salir")

    op = int(input("Seleccione: "))

    if op == 1:
        lista.insertarInicio(int(input("Dato: ")))

    elif op == 2:
        lista.insertarFinal(int(input("Dato: ")))

    elif op == 3:
        lista.insertarOrdenado(int(input("Dato: ")))

    elif op == 4:
        print("Encontrado:", lista.buscar(int(input("Buscar: "))))

    elif op == 5:
        lista.eliminar(int(input("Eliminar: ")))

    elif op == 6:
        lista.mostrar()

    elif op == 7:
        lista.ordenarAsc()
        print("Lista ordenada ascendentemente")

    elif op == 8:
        lista.ordenarDesc()
        print("Lista ordenada descendentemente")

    elif op == 0:
        print("Fin del programa")
        break

    else:
        print("Opción inválida")
```
